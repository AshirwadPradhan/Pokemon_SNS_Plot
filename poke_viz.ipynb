```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# importing seaborn
import seaborn as sns
```

```python
# import the dataset
dataset = pd.read_csv('data/Pokemon.csv')
print(dataset.head())
```

```python
# plotting scatter plot
sns.lmplot(x='Attack', y='Defense', data=dataset)
```

```python
# plotting scatter plot
# without regression line
# adding third dimension by coloring point based on generation
sns.lmplot(x='Attack', y='Defense', data=dataset, 
           fit_reg=False, hue='Generation')
```

```python
# limit the axis of the plot using matplotlib
sns.lmplot(x='Attack', y='Defense', data=dataset,
          fit_reg=False, hue='Generation')
#tweaks
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# boxplot using seaborn
# all the boxplots
sns.boxplot(data=dataset)
```

```python
# manipulating dataframe to make boxplots
# removing #, Total, Generation, Legendary
stats_df = dataset.drop(['#', 'Total', 'Generation', 'Legendary'], axis=1)
sns.boxplot(data=stats_df)
```

```python
# manipulating dataframe to make boxplots
# removing #, Total, Generation, Legendary
stats_df = dataset.drop(['#', 'Total', 'Generation', 'Legendary'], axis=1)
sns.boxplot(data=stats_df)
# setting themes in seaborn
sns.set_style('whitegrid')
```

```python
# violin plot(Type1 vs Attack)
sns.violinplot(x='Type 1', y='Attack', data=dataset)
```

```python
# swarm plot
sns.swarmplot(x='Type 1', y='Attack', data=dataset)
#plt.xlim(0,None)
```
